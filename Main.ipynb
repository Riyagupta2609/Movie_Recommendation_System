```python
movies=movies[['genre_ids','id','overview','title']]
movies
```

```python
import ast

ast.literal_eval("[18,80]")
```

```python
genre_map = {
    28: "Action",
    12: "Adventure",
    16: "Animation",
    35: "Comedy",
    80: "Crime",
    99: "Documentary",
    18: "Drama",
    10751: "Family",
    14: "Fantasy",
    36: "History",
    27: "Horror",
    10402: "Music",
    9648: "Mystery",
    10749: "Romance",
    878: "SciFi",
    10770: "TVMovie",
    53: "Thriller",
    10752: "War",
    37: "Western"
}
```

```python
def convert_genres(x):

    ids = ast.literal_eval(x)

    names = []

    for i in ids:
        if i in genre_map:
            names.append(genre_map[i])

    return " ".join(names)
```

```python
movies["genres"] = movies["genre_ids"].apply(convert_genres)
```

```python
movies["Tags"] = (
    movies["genres"]
    + " " + movies["overview"]
)
```

```python
movies
```

```python
new_data=movies.drop(columns=['overview','genres'])
```

```python
new_data
```

```python
movies.duplicated().sum()
```

```python
movies.drop_duplicates(inplace=True)
```

```python
movies.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv= CountVectorizer(max_features=10000,stop_words='english')
```

```python
vector= cv.fit_transform(new_data['Tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity= cosine_similarity(vector)
similarity
```

```python
new_data[new_data['title']=='The Godfather'].index[0]
```

```python
new_data.info()
```

```python
distance=sorted(list(enumerate(similarity[1])),reverse=True,key=lambda x:x[1])
for i in distance[0:6]:
  print(new_data.iloc[i[0]].title)
```

```python
def recommend(movies):
  index=new_data[new_data['title']==movies].index[0]
  distance=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
  for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
recommend("Iron Man")
```
